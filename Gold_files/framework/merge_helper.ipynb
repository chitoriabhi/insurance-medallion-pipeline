```python
from delta.tables import DeltaTable

def execute_merge(
    spark,
    source_df,
    target_table,
    merge_condition,
    update_map,
    insert_map
):

    target = DeltaTable.forName(
        spark,
        target_table
    )

    (
        target.alias("t")
        .merge(
            source_df.alias("s"),
            merge_condition
        )
        .whenMatchedUpdate(
            set=update_map
        )
        .whenNotMatchedInsert(
            values=insert_map
        )
        .execute()
    )
```
