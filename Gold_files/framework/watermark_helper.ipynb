```python
def get_watermark(table_name):

    result = spark.sql(f"""
        SELECT COALESCE(
            MAX(processed_ts),
            TIMESTAMP('1900-01-01')
        ) AS watermark
        FROM audit.gold_watermarks
        WHERE table_name = '{table_name}'
    """)

    return result.collect()[0]["watermark"]


def update_watermark(table_name):

    spark.sql(f"""
        INSERT INTO audit.gold_watermarks
        VALUES (
            '{table_name}',
            current_timestamp()
        )
    """)
```
