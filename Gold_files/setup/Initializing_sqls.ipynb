```python
%sql
CREATE SCHEMA IF NOT EXISTS audit;

CREATE TABLE IF NOT EXISTS audit.gold_watermarks (
    table_name STRING,
    processed_ts TIMESTAMP
)
USING DELTA;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS gold;

CREATE TABLE IF NOT EXISTS gold.dim_customer (
    customer_id STRING,
    first_name STRING,
    last_name STRING,
    email STRING,
    city STRING,
    state STRING,
    gold_updated_ts TIMESTAMP
)
USING DELTA;
```

```python
%sql
CREATE TABLE IF NOT EXISTS gold.fact_claim (
    claim_id STRING,
    customer_id STRING,
    policy_id STRING,
    claim_amount DECIMAL(18,2),
    claim_status STRING,
    claim_date DATE,
    claim_year INT,
    gold_updated_ts TIMESTAMP
)
USING DELTA
PARTITIONED BY (claim_year);
```
